```python
import numpy as np
from numpy.linalg import eig
import matplotlib.pyplot as plt
```

```python
a = np.array([[9, 4], 
              [3, 4]])
```

```python
plt.rcParams["figure.figsize"] = [5, 5]
origin = np.array([[0, 0], [0, 0]])
plt.quiver(*origin, a[0, 0], a[1, 0], scale=20, color=['black'])
plt.quiver(*origin, a[0, 1], a[1, 1], scale=20, color=['red'])
plt.grid()
plt.show()
```

```python
val, vec = eig(a)
print('E-value:', val)
print('E-vector', vec)
```
